```python
# 

# 问题
# 读入文档：
## https://blog.csdn.net/weixin_49779629/article/details/118297616
## https://blog.csdn.net/weixin_43238031/article/details/119970433

## drop数据：https://blog.csdn.net/weixin_55579895/article/details/122730654

import xlrd
import pandas as pd
from openpyxl import load_workbook
target=pd.DataFrame()



excel_name=r'C:\Users\liumeishan1\analysis\merge_sheet\daily.xls'
wb=xlrd.open_workbook(excel_name)

sheets=wb.sheet_names()
alldata=pd.DataFrame()
for i in range(len(sheets)-1):
    j=i+1
    df=pd.read_excel(excel_name,sheet_name=j,index_col=False)
    alldata=alldata.append(df)
    
    
alldata['日期']=alldata['日期'].map(str)
alldata['日期']=alldata['日期'].str.split(' ', expand=True)[0]


alldata['媒体']=alldata['媒体'].fillna('无')
alldata=alldata.drop(alldata[alldata['媒体']=='无'].index)


alldata['渠道']=alldata['媒体'].str.split('-', expand=True)[0]
alldata['端']=alldata['媒体'].str.split('-', expand=True)[1]

filtered_data=alldata.drop(alldata[alldata['日期'].str.contains(pat=r'.*总.*',regex=True)].index)
filtered_data=alldata.drop(alldata[alldata['媒体'].str.contains(pat=r'.*安卓.*',regex=True)].index)
#filtered_data=filtered_data.drop(filtered_data[filtered_data['端']=='汇总'].index)
#filtered_data=filtered_data.drop(filtered_data[filtered_data['端'].isin(['安卓','IOS'])].index)
print(filtered_data.info())

# get need cols
target['渠道']=filtered_data['渠道']
target['端']=filtered_data['端']
target['日期']=filtered_data['日期']
target['合计消耗']=filtered_data['合计消耗']
target['实际消耗']=filtered_data['实际消耗']
target['激活']=filtered_data['激活']
target['注册数']=filtered_data['注册数']
target['首日付费人数']=filtered_data['首日付费人数']
target['次留']=filtered_data['次留']
target['至今付费']=filtered_data['至今付费']
target['f1']=filtered_data['f1']
target['f3']=filtered_data['f3']
target['f7']=filtered_data['f7']
target['f15']=filtered_data['f15']
target['f30']=filtered_data['f30']
target['f60']=filtered_data['f60']
target['f90']=filtered_data['f90']
target['f180']=filtered_data['f180']
target['媒体']=filtered_data['媒体']

writer=pd.ExcelWriter(r'C:\Users\liumeishan1\analysis\merge_sheet\output.xlsx')
target.to_excel(writer,'target',index=False)
alldata.to_excel(writer,'all',index=False)
filtered_data.to_excel(writer,'filtered',index=False)
writer.save()

print('done')
```
