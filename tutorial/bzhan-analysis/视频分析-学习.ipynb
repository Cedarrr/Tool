```python
df=pd.read_excel(r'b站科技区2019-1~2020-3.xlsx')
df.head()
```

```python
# 汇总null的数据
df_null=df.isnull().sum()
df_null
```

```python
# 查找空值
df_null=round(df.isnull().sum()/len(df),4)
df_null
```

```python
# 剔除空值
df=df.dropna()
df.info()
```

```python
# ? 如何查看duplicated 的字段,pandas中如何定义duplicate这个东西
df[df.duplicated()].to_excel("duplicated.xlsx")
df=df.drop_duplicates()
df.info()
```

```python
# 把字符转换成时间戳
# ? 这里format是说原时间的格式，还是要转换成的格式
df['date']=pd.to_datetime(df['date'],format='%Y-%M-%D %H:%M')
df.info()
df.head()
```

```python
df.to_excel("video_info.xlsx",engine="xlsxwriter")
```

```python
df=df[['分区','bv','author','date','coins','danmu','favorite','likes','replay','share','view']]
df.head()
```

```python
# df loc和iloc区别在哪里
# https://www.yisu.com/zixun/176924.html
sc=df.loc[df['分区']=="科学科普"]
sc.head()
```

```python
# reset_index()用法
# TO DO -- 一会去测试一下效果
count_no_index=sc.groupby('author')['date'].count()
count=sc.groupby('author')['date'].count().reset_index()
count
count.info()
```

```python
# 发布大于5条视频的作者只有208个
count.columns=['author','Times']
com_m=count[count['Times']>5]
com_m.info()
```

```python
# datetime.dt.days 保留完整天数
last=sc.groupby('author')['date'].max()
late=sc.groupby('author')['date'].min()
F=round((last-late).dt.days/sc.groupby('author')['date'].count()).reset_index()
F.columns=['author','F']
F=pd.merge(com_m,F,on='author',how='inner')
F.head()
```
